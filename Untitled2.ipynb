```python
from google.colab import files
from PyPDF2 import PdfReader
upload=files.upload()
pdf_path=list(upload.keys())[0]
reader=PdfReader(pdf_path)
all_text=""
for page in reader.pages:
  text=page.extract_text()
  all_text=all_text+text+" "
print(all_text)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
splitter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=100)
chunks=splitter.split_text(all_text)
print(len(chunks))
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
embeddings=HuggingFaceEmbeddings(model_name="sentence-transformers/all-miniLM-L6-v2")
```

```python
from langchain_community.vectorstores import Chroma
db=Chroma.from_texts(chunks,embedding=embeddings,persist_directory="/content/pdf_chat")
db.persist()
print("My db is created and embedding vector stored successfully")
```

```python
query="hi tell me about uniform cost search"
results=db.similarity_search(query,k=3)
print(results)
```
